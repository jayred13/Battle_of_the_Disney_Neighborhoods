##### Scrape WIki

```python
""" Accesses webpage, searches for table, converts/returns panda dataframe """
def scrape_wiki(wiki_url, lang, park):
    
    """ Find table of park sections and attractions """
    def find_attractions_table( soup ): 
        tables = soup.find_all('div', class_='navbox')
        for i in range(len(tables)): 
            if tables[i]['aria-labelledby'] in park: return(tables[i])
    
    """ Fetch all park sections """
    def get_park_sections( soup ): return( [section.text for section in soup.find_all('span', class_='toctext')] )
    def get_section_name( header ): 
        if header.find('a'): #Hyperlink
            if header.find('a').text in park_sections: return( header.find('a').text ) 
        else: #No Hyperlink
            if header.text in park_sections: return( header.text ) 
        
    """ Find attraction in park section """
    def get_attractions( list_items ): return( [get_attraction_name(item) for item in list_items ] )
    def get_attraction_name( item ): 
        if item.find('a'): return( item.find('a').text ) #Hyperlink
        else: return( item.text ) #No Hyperlink
    
    """ Converts table into pandas dataframe """
    def create_dataframe( table ):
        df=pd.DataFrame(columns={'Attraction', 'Section', 'Park'})
        all_sections=[] 
        all_attractions=[]
        
        for row in table.find_all('tr')[2:]:
            header = row.find('th')
            section_name = get_section_name(header)  
            
            list_items = row.find_all('li')
            attractions = get_attractions(list_items)  
            
            all_sections.extend( [section_name]*len(attractions) )
            all_attractions.extend( attractions )
            
        df['Attraction'] = all_attractions 
        df['Section'] = all_sections
        df['Park'] = park
        df.drop_duplicates(subset ="Attraction", keep = 'first', inplace = True) 
        return( df.dropna(subset=['Section']) )
        
    """ Main calls """
    response = requests.get(wiki_url)
    soup = BeautifulSoup(response.content, lang)
    
    park_sections = get_park_sections( soup )
    print(f"{park} Sections:\n{park_sections}")
    
    table = find_attractions_table(soup)    
    return( create_dataframe(table) )
```

### Dataset #2.1 - Attraction Names and Wait Times

#### Methods

##### Scrape Current Wait TIme Table

```python
""" Accesses webpage, searches for table, converts/returns panda dataframe """
def scrape_current_table(webpage_url, lang):
    
    """ Rows w/ Icons display wait times differently"""
    def get_current_wait( row ):
        row_data = [i.text for i in row.find_all('td')]
        
        if row.find('table'): # Row has table/icon
                wait = row.find('table').find_all('td')[0]
                return( wait.string )
        else:
            wait = row_data[1]
            return( wait.split('F')[0] ) # Removes Fast-Pass info
                
    """ Rows w/ Hyperlinks display names differently"""
    def get_name( row ):
        if row.find('a'): return( row.find('a').text ) # Hyperlink
        else: return( row.find_all('td')[0].string ) # No hyperlink
    
    """ Find hyperlink within a givien row """
    def get_link( row ):
        if row.find('a'): return( row.find('a')['href'] )
    
    """ Converts table into pandas dataframe """
    def create_dataframe( table ):
        attraction_names = []
        wait_times = []
        links = []
        
        for row in table:
            if row.find('span'): # Filters only displayed rows
                #wait_times.append( get_current_wait(row) )
                attraction_names.append( get_name(row) )
                links.append( get_link(row) )
                
        df = pd.DataFrame()
        df['Attraction'] = attraction_names
        df['Wait Time'] = wait_times
        df['URL'] = links
        
        return( df.dropna(subset=['URL']) )
     
    """ Main Calls """
    response = requests.get(webpage_url)
    soup = BeautifulSoup(response.content, lang)
    
    table = soup.find('table').find_all('tr') # Get Table
    return( create_dataframe(table) )
```

##### Scrape Average Wait TIme Table

```python
""" Accesses webpage, searches for table, converts/returns panda dataframe """
def scrape_average_table(webpage_url, lang):
    
    """ Rows w/ Icons display wait times differently"""
    def get_average_wait( row ):
        #print(row.find_all('td')[2].text)
        return( row.find_all('td')[2].text )
    
    """ Rows w/ Hyperlinks display names differently"""
    def get_name( row ):
        if row.find('a'): 
            #print(row.find('a').text)
            return( row.find('a').text ) # Hyperlink
        else: return( row.find_all('td')[0].string ) # No hyperlink
        
    def get_link( row ):
        if row.find('a'): 
            #print(row.find('a')['href'])
            return( row.find('a')['href'] )
    
    def create_dataframe( table ):
        attraction_names = []
        wait_times = []
        links = []
        
        for row in table:            
            wait_times.append( get_average_wait(row) )
            attraction_names.append( get_name(row) )
            links.append( get_link(row) )
                
        df = pd.DataFrame()
        df['Attraction'] = attraction_names
        df['Wait Time'] = wait_times
        df['URL'] = links
        return( df.dropna(subset=['URL']) )
                
    response = requests.get(webpage_url)
    soup = BeautifulSoup(response.content, lang)
    
    table = soup.find_all('table')[0].find_all('tr')[1:] # Get Table
    #print(table)
    
    return( create_dataframe(table) )
```

##### Get Attraction Details

```python
""" Collects Attraction names by calling helper methods, returns panda dataframe """
def get_disney_attractions(location, wait_tense):
        
    """ Getting ride name, wait times, & links """    
    def get_attractions(park):
        if wait_tense == 'current_wait':
            attractions = scrape_current_table(f"https://www.laughingplace.com/w/p/{park}-current-wait-times/", 'lxml')
            return(attractions)
        
        elif wait_tense == 'average_wait':
            attractions = scrape_average_table(f"https://www.laughingplace.com/w/p/{park}-wait-time-averages-for-last-30-days/", 'lxml')
            return(attractions)
        
    if location == 'California':
        if wait_tense == 'current_wait':
            disland_attractions = get_attractions('disneyland')
            print('Disneyland Attractions: ', disland_attractions.shape)

            caladv_attractions = get_attractions('disney-california-adventure')
            print('Californa Advenutre Atractions: ', caladv_attractions.shape)

            all_attractions = disland_attractions.append( caladv_attractions, ignore_index=True )
            all_attractions.sort_values('Attraction') 
            all_attractions.drop_duplicates(subset = 'Attraction', keep = 'first', inplace = True)
            print('All (California) Disney Attractions: ', all_attractions.shape)
        
        elif wait_tense == 'average_wait':
            all_attractions = get_attractions('disneyland-resort')
        
        return( all_attractions )
        
    else:
        print("Error: Cannot find any wait times.")
        return( pd.DataFrame() )
```

##### Get Attraction Details

```python
""" Grabs details from individual ride webpages """
def get_attraction_info(attractions):
    
    """ Helper methods for collecting information """
    def get_attraction_type(soup):
        if soup.find('td'):
            return( soup.find_all('td')[1].string )
    def get_attraction_length(soup): 
        if soup.find('td'):
            return( soup.find_all('td')[3].string )
    def get_attraction_wet_factor(soup): 
        if soup.find('td'):
            return( soup.find_all('td')[5].string )
    def get_attraction_height(soup): 
        if soup.find('td'):
            return( soup.find_all('td')[7].string )
    def get_attraction_popularity(soup): 
        if soup.find('td'):
            return( soup.find_all('td')[9].string )
    def get_attraction_description(soup): 
        if soup.find('td'):
            return( soup.find('div', class_="et_pb_text_inner").find('p').string  )
    
    """ Main Calls """  
    print("\nGetting Attraction details. This may take several minutes...")  
    numRows = len(attractions.index)
    attraction_info = [[] for _ in range(6)]
    
    print("# of Attractions remaining: ")
    for row in attractions.index:
        url = attractions['URL'][row]
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html')
        
        attraction_info[0].append( get_attraction_type( soup ) )
        attraction_info[1].append( get_attraction_length( soup ) )
        attraction_info[2].append( get_attraction_wet_factor( soup ) )
        attraction_info[3].append( get_attraction_height( soup ) )
        attraction_info[4].append( get_attraction_popularity( soup ) )
        attraction_info[5].append( get_attraction_description( soup ) )
        
        #print(attractions['Attraction'][row], attraction_type)
        print(numRows, end ="...")
        numRows-=1
        
    attractions['Type'] = attraction_info[0]
    attractions['Length'] = attraction_info[1]
    attractions['Wet Factor'] = attraction_info[2]
    attractions['Height Require'] = attraction_info[3]
    attractions['Popularity'] = attraction_info[4]
    attractions['Description'] = attraction_info[5]
        
    return( attractions.dropna(subset=['Type']) )
```

### Merging Datasets

#### Methods

```python
""" Merges attraction datasets based on the name of the attraction  """
def merge_datasets( df, df2 ):
    
    """ Creates a column of easily comparable attraction names """
    def raw_text( df ):
        df['raw_text'] = df['Attraction']
        df['raw_text'] = df['raw_text'].str.lower() 
        df['raw_text'] = df['raw_text'].str.replace('[^\w\s]','') 
        df['raw_text'] = df['raw_text'].str.replace('\s+', '_', regex=True)
        return(df)
    
    def print_it( df ):
        for index, row in df.iterrows():
            print(index, row['Attraction'])
    
    """ Merges 2 datasets after fixing duplicates """
    def merge( df, df2 ):
        """ Finds & fixes names that are a variation of each other """
        for a_index, a_row in df.iterrows():
            for b_index, b_row in df2.iterrows():
                if a_row['raw_text'] in b_row['raw_text']: 
                    a_row['Attraction'] = b_row['Attraction']
                    
                elif b_row['raw_text'] in a_row['raw_text']:
                    b_row['Attraction'] = a_row['Attraction']
        
        return( pd.merge(df2, df) )
     
    """ Main Calls """    
    df = raw_text( df )
    df2 = raw_text( df2 )
    df_combo = merge( df, df2 )
    df_combo = df_combo.drop(['raw_text'], axis=1)
    return( df_combo )
```

### K-means Attraction Clustering Class

```python
class kmeans_cluster:
    df = pd.DataFrame()
    col_name = ''
    num_clusters = 0 
    
    def __init__(self, df, col_name, num_clusters):
        self.df = df
        self.col_name = col_name
        self.num_clusters = num_clusters
    
    """ Performs one-hot-encoding on a given dataframe, provided a column name """
    def one_hot_encode( df, col_name ):
        df_onehot =  pd.get_dummies( df[[col_name]], prefix="", prefix_sep="" )
        df_onehot['Section'] = df['Section']
        return(df_onehot)
    
    """ Returns frequency of the occurence of each category """
    def cat_freq( df_onehot ):
        df_grouped = df_onehot.groupby('Section').mean().reset_index()
        return(df_grouped)
    
    """ Helper method that returns most common items for a given section """
    def return_most_common_items(row, num_top_items):
        row_categories = row.iloc[1:]
        row_categories_sorted = row_categories.sort_values(ascending=False)
        return row_categories_sorted.index.values[0:num_top_items]
    
    """ Orders most common item per section """
    def order_most_common( df_grouped, num_top_items ):
        indicators = ['st', 'nd', 'rd']

        # create columns according to number of top venues
        columns = ['Section']
        for ind in np.arange(num_top_items):
            try:
                columns.append('{}{} Most Common...'.format(ind+1, indicators[ind]))
            except:
                columns.append('{}th Most Common...'.format(ind+1))

        # create a new dataframe
        df_items_sorted = pd.DataFrame(columns=columns)
        df_items_sorted['Section'] = df_grouped['Section']

        for ind in np.arange(df_grouped.shape[0]):
            df_items_sorted.iloc[ind, 1:] = kmeans_cluster.return_most_common_items(df_grouped.iloc[ind, :], num_top_items)

        return( df_items_sorted )
    
    """ Performs the clustering algorithm on the given dataframe """
    def cluster( df, df_grouped, df_items_sorted , num_clusters ):
        df_grouped_clustering = df_grouped.drop('Section', 1)

        # run k-means clustering
        kmeans = KMeans(num_clusters, random_state=0).fit(df_grouped_clustering)

        # check cluster labels generated for each row in the dataframe
        kmeans.labels_[0:10] 

        # add clustering labels
        df_items_sorted.insert(0, 'Cluster Labels', kmeans.labels_, allow_duplicates=True)

        df_merged = df.copy()

        # merge df_grouped with df to add latitude/longitude
        df_merged = df_merged.join(df_items_sorted.set_index('Section'), on='Section')
        return(df_merged)
    
    def run(self):
        df_onehot = kmeans_cluster.one_hot_encode( self.df, self.col_name )
        df_grouped = kmeans_cluster.cat_freq( df_onehot )
        df_items_sorted =  kmeans_cluster.order_most_common( df_grouped, 5 )
        df_merged = kmeans_cluster.cluster( self.df, df_grouped, df_items_sorted, self.num_clusters)
        return(df_merged)
```
